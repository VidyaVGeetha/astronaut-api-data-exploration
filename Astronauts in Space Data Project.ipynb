# Project Name: Exploring Public APIs: Astronauts in Space Data Project

```python
# Step 1: Import required libraries.
import requests # you need this module to make an API call
import pandas as pd
```

```python
# Step 2: Store the Astronaut API URL in a Variable
api_url="http://api.open-notify.org/astros.json"
```

```python
# Step 3: Fetch Data from the API
response= requests.get(api_url)   # Call the API using the get method and store the
                                # output of the API call in a variable called response
```

```python
# Step 4: Check if the API Request Was Successful and Read the JSON Data
if response.ok:          # if all is well() no errors, no network timeouts)
    data=response.json() # store the result in json format in a variable called data
                            # the variable data is of type dictionary.
```

```python
# Step 5:  print the data just to check the output or for debugging
print(data)
```

Note:

   
**API Response (Astronaut Data):**  
The API returned a JSON object containing:
- `number` → Total number of people currently in space
- `people` → A list of astronauts, including their name and the spacecraft they are on
- `message` → Status of the API request (success)

In this result:
- Total astronauts in space = 12
- Most astronauts are on the ISS, and some are on the Tiangong spacecraft.


```python
# Step 6: Print the number of astronauts currently on ISS.
print(data.get('number'))
```

```python
# Step 7: Display the Number of Astronauts and Their Names
astronauts= data.get('people')
print("There are {} astronauts on ISS".format(len(astronauts)))
print("And their names are :")
for astronaut in astronauts:
    print(astronaut.get('name'))
```
